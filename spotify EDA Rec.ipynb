```python
# Ensure all columns used for correlation are numeric
correlation_columns = ['popularity', 'duration_ms', 'explicit', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']

# Convert all columns to numeric, coercing errors to NaN (which will then be ignored in correlation calculation)
df_tracks[correlation_columns] = df_tracks[correlation_columns].apply(pd.to_numeric, errors='coerce')

# Recalculate the correlation matrix for the numerical features of the dataset
correlation_matrix = df_tracks[correlation_columns].corr()

# Extract the 'popularity' column's correlation values
popularity_correlation = correlation_matrix['popularity'].sort_values(ascending=False)

# Display the correlation values with 'popularity'
print(popularity_correlation)
```

```python
# Convert release dates to datetime objects and extract the year
tracks_df['release_date'] = pd.to_datetime(tracks_df['release_date'], errors='coerce')
tracks_df['year'] = tracks_df['release_date'].dt.year

# Aggregate popularity by year
popularity_trend = tracks_df.groupby('year')['popularity'].mean().reset_index()

# Sort the results by year
popularity_trend = popularity_trend.sort_values(by='year')

# Plot the results to visualize the trend
plt.figure(figsize=(14,7))
plt.plot(popularity_trend['year'], popularity_trend['popularity'], marker='o')
plt.title('Average Popularity of Songs Over Time')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(True)
plt.show()
```
